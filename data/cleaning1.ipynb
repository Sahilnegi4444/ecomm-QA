```python
import json

data = []
with open ('C:\Projects\ecomm qa\data\Beauty_and_Personal_Care.jsonl', 'r') as f:
    for line in f:
        data.append(json.loads(line))
```

```python
import pandas as pd

df = pd.DataFrame(data)
```

```python
df.head()
```

```python
df.drop(columns=['images', 'asin', 'user_id', 'timestamp','timestamp','helpful_vote','verified_purchase'], inplace=True)
```

```python
df.head()
```

```python
import json

meta_data = []
with open ('C:\Projects\ecomm qa\data\meta_Beauty_and_Personal_Care.jsonl', 'r') as f:
    for line in f:
        meta_data.append(json.loads(line))
```

```python
meta_df = pd.DataFrame(meta_data)
meta_df.head(3)
```

```python
print(meta_df.shape)
print(df.shape)
```

```python
print(df.dtypes['parent_asin'])
print(meta_df.dtypes['parent_asin'])
```

```python
merged_df = pd.merge(df, meta_df, on='parent_asin', how='inner')
```

```python
merged_df.sample(3)
```

```python
merged_df.drop(columns = ['bought_together','subtitle','author','videos','images'],inplace = True)
```

```python
merged_df.head()
```

```python
merged_df.isnull().sum()
```

```python
merged_df.shape
```

```python
missing_by_category = merged_df[merged_df['price'].isna()].groupby('main_category').size()
missing_by_category
```

```python
remaining_by_category = merged_df[merged_df['price'].notna()].groupby('main_category').size()
remaining_by_category
```

```python
amazon_df = merged_df
amazon_df.head()
```

```python
amazon_df = amazon_df.dropna(subset=['price'])              #drop rows based on the missing price values
```

```python
amazon_df.isnull().sum()                
```

```python
#converting the list and dict to push into db
columns_to_convert = ['features', 'categories', 'details', 'description']

for col in columns_to_convert:
    if col in amazon_df.columns:    
        amazon_df[col] = amazon_df[col].apply(
            lambda x: json.dumps(x) if isinstance(x, (dict, list)) else x
        )
```

```python
for col in columns_to_convert:
    if col in amazon_df.columns:
        sample = amazon_df[col].iloc[0] if len(amazon_df) > 0 else None
        print(f"{col}: {type(sample).__name__}")
```

```python
amazon_df['parent_asin'].duplicated().sum()
```

```python
from sqlalchemy.types import Integer, String, Float, Text
dtype_dict = {
    'rating': Float,
    'review_title': Text,
    'review_text': Text,
    'parent_asin': String(20),
    'category': String(100),
    'product_title': Text,
    'average_rating': Float,
    'rating_number': Integer,
    'features': Text,
    'description': Text,
    'price': Float,
    'store': String(200),
    'categories': Text,
    'details': Text
}
```

```python
amazon_df.shape
```

```python
print(f"Original: {len(amazon_df):,}")
amazon_df = amazon_df.iloc[::4].reset_index(drop=True)          #Taking every fourth row to reduce rows
print(f"After reducing by 3: {len(amazon_df):,}")
```

```python
from sqlalchemy import create_engine


DATABASE_URL = 'mysql+pymysql://root:Lihasigen%404444@localhost:3306/amazon_data_db'

engine = create_engine(DATABASE_URL)
```

```python
try:
    amazon_df.to_sql(
        name='amazon_dataframe', #sql table
        con=engine,
        if_exists='replace',        
        index=False,                
        chunksize=10000,            #Insert in batches
        method='multi',              #bulk insert statements for speed
        dtype=dtype_dict
    )
    print("Data succesfully ingested")

except Exception as e:
    print(f"An error occurred: {e}")

```
