####  There are not much thing to clean, I'll just change the name of some of the columns and remove list and dict patterns 

```python
df = df.rename(columns={'text':'reviews','title_x':'review_headline','title_y':'review_title','main_category':'category'})
df.head()
```

```python
# removing list patterns
df['features'] = df['features'].str.replace(r"[\[\]\"']", '', regex=True)
df['description'] = df['description'].str.replace(r"[\[\]\"']", '', regex=True)
df['categories'] = df['categories'].str.replace(r"[\[\]\"']", '', regex=True)
```

```python
# removing dict patterns
df['details'] = df['details'].str.replace(r'[{}"\']', '', regex=True)
```

```python
df.head()
```

```python
df = df.rename(columns={'review_title':'product_title'})
```
